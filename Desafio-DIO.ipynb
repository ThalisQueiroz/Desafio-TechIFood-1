## Objetivo

Crie uma variável para armazenar o nome e a quantidade de experiência (XP) de um herói, depois utilize uma estrutura de decisão para apresentar alguma das mensagens abaixo:

Se XP for menor do que 1.000 = Ferro
Se XP for entre 1.001 e 2.000 = Bronze
Se XP for entre 2.001 e 5.000 = Prata
Se XP for entre 6.001 e 7.000 = Ouro
Se XP for entre 7.001 e 8.000 = Platina
Se XP for entre 8.001 e 9.000 = Ascendente
Se XP for entre 9.001 e 10.000= Imortal
Se XP for maior ou igual a 10.001 = Radiante

## Saída

Ao final deve se exibir uma mensagem:
"O Herói de nome **{nome}** está no nível de **{nivel}**"

```python
while True:
    nome = input("Digite o nome do herói (ou 'sair' para encerrar): ")

    if nome == 'sair':
        print("Encerrando o programa.")
        break

    quantidadeXP = int(input("Digite a quantidade de experiência do herói: "))

    if quantidadeXP < 1000:
        nivel = "Ferro"
    elif 1001 <= quantidadeXP <= 2000:
        nivel = "Bronze"
    elif 2001 <= quantidadeXP <= 5000:
        nivel = "Prata"
    elif 5001 <= quantidadeXP <= 7000:
        nivel = "Ouro"
    elif 7001 <= quantidadeXP <= 8000:
        nivel = "Platina"
    elif 8001 <= quantidadeXP <= 9000:
        nivel = "Ascendente"
    elif 9001 <= quantidadeXP <= 10000:
        nivel = "Imortal"
    else:
        nivel = "Radiante"

    print(f"O Herói de nome {nome} está no nível de {nivel}")
```
